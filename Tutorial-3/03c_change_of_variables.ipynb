```python
import matplotlib.pyplot as plt
import numpy as np

seed = 42
np.random.seed(seed)
```

## <u>Question 1b</u>

```python
mu = 1
sigma = 2

# define a range to calculate the PDF of Y:
y_range = np.arange(1, np.exp(2*sigma), 0.1)
# compute the PDF with the formula that we found analytically:
pdfy = np.exp( -(np.log(y_range) - mu)**2 / (2*sigma**2) ) / (y_range*sigma*np.sqrt(2*np.pi))

fig, ax = plt.subplots(1, 3, figsize=(14, 4))
for idx_n, n in enumerate([100, 1_000, 10_000]):

    # n samples from a Gaussian with mean 1 and variance 4:
    x = sigma*np.random.randn(n) + mu
    # discard samples outside the 96% (2*sigma) CI for better visualization:
    x = x[np.abs(x) < 2*sigma]
    # exponentiate the samples:
    y = np.exp(x)
    # plot:
    ax[idx_n].hist(y, bins=30, density=True, label='Histogram')
    ax[idx_n].plot(y_range, pdfy, label='PDF')
    ax[idx_n].legend(loc="upper right")
    ax[idx_n].set_xlim([0, 60])
    ax[idx_n].set_title(f'n={n}')

# plt.savefig('data/ex1b.pdf', bbox_inches='tight', pad_inches=0)
plt.show()
plt.close()
```

## <u>Question 1d</u>

```python
lamda1 = 1
lamda2 = 2

# define a range to calculate the pdf of Y:
y_range = np.arange(0.1, 10, 0.1)
# compute the PDF with the formula we found analytically:
pdfy = (lamda1*lamda2)/(lamda1 - lamda2) * (np.exp(-lamda2*y_range) - np.exp(-lamda1*y_range))

fig, ax = plt.subplots(1, 3, figsize=(14, 4))
for idx_n, n in enumerate([100, 1_000, 10_000]):

    # n samples uniform in (0, 1):
    u1 = np.random.rand(n)
    u2 = np.random.ranf(n)
    y = -np.log(u1)/lamda1 -np.log(u2)/lamda2
    # plot:
    ax[idx_n].hist(y, bins=30, density=True, label='Histogram')
    ax[idx_n].plot(y_range, pdfy, label='PDF')
    ax[idx_n].legend(loc="upper right")
    ax[idx_n].set_ylim([0, 0.8])
    ax[idx_n].set_xlim([0, 10])
    ax[idx_n].set_title(f'n={n}')

# plt.savefig('data/ex1d.pdf', bbox_inches='tight', pad_inches=0)
plt.show()
plt.close()
```
